## Data Beschrijving

Inzicht in het calorieverbruik tijdens fysieke activiteit is belangrijk om de relatie tussen beweging, gezondheid en calorieverbruik beter te begrijpen. Deze dataset biedt gedetailleerde informatie over individuen en hun lichaamskenmerken, gecombineerd met gegevens over hun fysieke inspanning. Het doel is om voorspellende modellen te ondersteunen die het aantal verbrande calorieën tijdens een workout kunnen inschatten.

De dataset bevat zowel persoonlijke kenmerken van de deelnemer (zoals leeftijd, lengte, gewicht en geslacht) als gegevens die rechtstreeks verband houden met de inspanning (zoals duur, hartslag en lichaamstemperatuur). Dit maakt het mogelijk om de invloed van zowel fysiologische factoren als trainingsintensiteit te analyseren.

1. **ID**: unieke identificatiecode voor elk datapunt in de dataset
1. **Sex**: het biologische geslacht van de persoon, doorgaans gecodeerd als man of vrouw
1. **Age**: de leeftijd van de persoon
1. **Height**: de lengte van de persoon
1. **Weight**: het gewicht van de persoon
1. **Duration**: de duur van de fysieke activiteit (in minuten)
1. **Heart_Rate**: de hartslag tijdens de activiteit (in slagen per minuut)
1. **Body_Temp**: de lichaamstemperatuur tijdens de activiteit
1. **Calories**: het totale aantal verbrande calorieën tijdens de fysieke activiteit (in kcal)
