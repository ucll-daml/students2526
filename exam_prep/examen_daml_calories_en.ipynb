## Data Description

Understanding calorie expenditure during physical activity is important for gaining better insight into the relationship between exercise, health, and energy use. This dataset provides detailed information about individuals and their physical characteristics, combined with data on their physical effort. The aim is to support predictive models that can estimate the number of calories burned during a workout.

The dataset contains both personal characteristics of the participant (such as age, height, weight, and sex) as well as data directly related to the exercise (such as duration, heart rate, and body temperature). This makes it possible to analyze the influence of both physiological factors and training intensity.

1. **ID**: unique identifier for each data point in the dataset
1. **Sex**: the biological sex of the person, typically coded as male or female
1. **Age**: the person’s age
1. **Height**: the person’s height
1. **Weight**: the person's weight
1. **Duration**: the duration of the physical activity (in minutes)
1. **Heart_Rate**: the heart rate during the activity (in beats per minute)
1. **Body_Temp**: the body temperature during the activity
1. **Calories**: the total number of calories burned during the physical activity (in kcal)
